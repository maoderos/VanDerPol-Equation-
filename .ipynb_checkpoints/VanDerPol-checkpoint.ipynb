# Van der Pol Equation

```python
import numpy as np
import matplotlib.pyplot as plt
from numba import njit
@njit
def fX(v):
    return v
@njit
def fV(x,v,u,w_d,w_o,F,t):
    return (u*(1 - x**2)*v - w_o**2*x + F*np.cos(w_d*t))
    #return (-(w_o**2)*x)
    
@njit(parallel=True)
def RK2(x_o,v_o,t_o,tf,dt,u,w_d,w_o,F):
    X = x_o
    V = v_o
    t = t_o + dt
    X_list = [x_o]
    V_list = [v_o]
    T_list = [t_o]
    while(t < tf):
        k1X = fX(V)
        k1V = fV(X,V,u,w_d,w_o,F,t)
        Xaux = X + k1X*dt/2
        Vaux = V + k1V*dt/2 
        k2X = fX(Vaux)
        k2V = fV(Xaux,Vaux,u,w_d,w_o,F, t+dt/2)
        X = X + k2X*dt
        V = V + k2V*dt
        #print(V)
        #print(X)
        X_list.append(X)
        V_list.append(V)
        T_list.append(t)
        t += dt
    return X_list, V_list, T_list

# --- Initial values -----
x_o = 2 # position
v_o = 0.1 # velocity 
u = 3 # mu constant
w_d = 4 # frequency of the external force
w_o = 1 # frequency
F = 22 # external force amplitude
t_o = 0 # initial time
tf = 300 # final time
dt = 0.001

X_list, V_list, T_list = RK2(x_o,v_o,t_o,tf,dt,u,w_d,w_o,F)

plt.figure()
#print(V_list)
plt.title("Phase space")
plt.xlabel("x")
plt.ylabel("v")
plt.plot(X_list, V_list)
plt.show()
plt.close()

plt.figure()
#print(V_list)
plt.title("Position vs Time")
plt.xlabel("x")
plt.ylabel("t")
plt.xlim(0,50)
plt.plot(T_list, X_list)
plt.show()
plt.close()
```

```python
# PHASE SPACE FOR MU < -2 
POINT = [[0.1,0.04],[-0.1,0.04],[0.1,-0.6],[-0.1,-0.6],[0.1,0.6],[-0.1,0.6]]
#POINT = [[1,0.04]]
u = -3 # mu constant
F= 0
tf = 30
plt.figure()
plt.title("Phase space: $\mu = -3$")
plt.xlabel("x")
plt.ylabel("v")
for i in POINT:
    #print(i[1])
    X_list, V_list, T_list = RK2(i[0],i[1],t_o,tf,dt,u,w_d,w_o,F)
    #print(V_list)
    plt.plot(X_list, V_list)
    
plt.show()
plt.close()
```

```python
# PHASE SPACE FOR MU > 2 
POINT = [[0.1,0.04],[-0.1,0.04],[0.1,-0.6],[-0.1,-0.6],[0.1,0.6],[-0.1,0.6]]
#POINT = [[3,0.04]]
u = 3 # mu constant
F= 0
tf = 30
plt.figure()
plt.title("Phase space: $\mu = 3$; $x_o = \pm 0.1$")
plt.xlabel("x")
plt.ylabel("v")
for i in POINT:
    #print(i[1])
    X_list, V_list, T_list = RK2(i[0],i[1],t_o,tf,dt,u,w_d,w_o,F)
    #print(V_list)
    plt.plot(X_list, V_list)
    
plt.show()
plt.close()
```

```python

F = np.arange(1,12,0.01)
# --- Initial values -----
x_o = 0.5 # position
v_o = 0.0 # velocity 
u = 4 # mu constant
w_d = 3 # frequency of the external force
w_o = 1 # frequency

t_o = 0 # initial time
tf = 500 # final time
dt = 0.1

plt.figure()
plt.title("Bifurcation diagram:")
plt.xlabel("x")
plt.ylabel("F")
for i in F:
    #print(i)
    F_list = []
    xList = []
    for j in range(200):
        X_list, V_list, T_list = RK2(x_o,v_o,t_o,tf,dt,u,w_d,w_o,i)
        x_o,v_o = X_list[-1], V_list[-1]
    for j in range(100):
        X_list, V_list, T_list = RK2(x_o,v_o,t_o,tf,dt,u,w_d,w_o,i)
        x_o,v_o = X_list[-1], V_list[-1]
        xList.append(X_list[-1])
        F_list.append(i)
    plt.scatter(F_list,xList,marker='.',s=1)
#plt.xlim(24,26)
plt.show()
```
